```python
#1.	Read the demand transaction file and print the first 15 rows and dtypes
import pandas as pd
url = 'https://raw.githubusercontent.com/balaji2v/Inceptez_Batch19/main/Demand_txn_updated.csv'
df = pd.read_csv(url)
df.dtypes
#print(df.to_string())
```

```python
#2.	Convert the transaction date column to a date column using pd.to_datetime function
print("Before : ",df['Transaction_Date'])
df['Transaction_Date'] = df['Transaction_Date'].apply(pd.to_datetime)
print("Aft : ",df['Transaction_Date'])
```

```python
#3.	Subset the dataframe for transaction date greater than '2016-08-01'
transDate_subset = df.loc[df['Transaction_Date'] > '2016-08-01']
print(transDate_subset)
```

```python
#4.	Look for unique values in Mapped_Sales_Type
print(df['Mapped_Sales_Type'].unique())
```

```python
#5.	Subset the entire dataframe, based on the below condition, to a new dataframe and work on the following questions
#a.	Condition: Avg_Discount_Percent_On_Discounted_Items should be less than 1.0

sub_df = df.loc[df['Avg_Discount_Percent_On_Discounted_Items']<1.0]
print(sub_df)
```

```python
#b.	Check for sanity if the new dataframe contains Avg_Discount_Percent_On_Discounted_Items greater than or equal to 1.0
print(sub_df.loc[df['Avg_Discount_Percent_On_Discounted_Items']>=1.0])
```

```python
"""6.	Groupby 'City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date' and perform following aggregate operations on respective columns as mentioned (note: Only one groupby to do all the below aggregations)
a.	Quantity_Sold – sum
b.	Median_Price – median
c.	Effective_Price – median"""
df.groupby(['City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date']).agg({'Quantity_Sold':'sum','Median_Price':'median','Effective_Price':'median'})
```

```python
#7.	Display the data for Effective_Price = not null and just print the last 5 rows (hint: use .notnull function)
Effective_Price_Subset = df.loc[df['Effective_Price'].notnull()]
print(Effective_Price_Subset.tail(5))
```

```python
#8.	Display the data for city = Chennai AND Mapped_Sales_Type = Delivery and print top 5 records (hint use loc function to subset, then “&” operator 
#to filter both the above cities)
City_SalesType_Subset = df.loc[(df['City'] == 'Chennai') & (df['Mapped_Sales_Type'] == 'Delivery')]
print(City_SalesType_Subset.head(5))
```

```python
#Display the data for the column “Day” with values Mon, Tue and Wed (hint: use .isin function)
day_subset = df.loc[df['Day'].isin(["Mon","Tue","Wed"])]
print(day_subset)
```

```python
#10.	Display the data for Percent_Quantity_With_Discount not equal to 0.0 (hint: loc function with != operator)
PerQuan_Subset = df.loc[(df['Percent_Quantity_With_Discount'] != 0.0)]
print(PerQuan_Subset)
```

```python
#11.	Add a column “Range” to the existing dataframe for below condition
#a.	1 for Effective_Price > 500
#b.	0.5 otherwise
Range_col = []
for i in df['Effective_Price']:
  if i>500: Range_col.append(1)
  else: Range_col.append(0.5)
df['Range'] = Range_col
print(df[['Effective_Price','Range']])
```

```python
#12.	Read the demand transaction file again and perform following operations for practice
#a.	Handle the missing values in the column Percent_Quantity_With_Discount with strategy = mean
#print(df['Percent_Quantity_With_Discount'].to_string())
df['Percent_Quantity_With_Discount'].fillna(df['Percent_Quantity_With_Discount'].mean(), inplace=True)
print(df['Percent_Quantity_With_Discount'])
```
